```python
import os
import librosa
import matplotlib.pyplot as plt
import numpy as np
import librosa.display
import IPython
import wavio
import sounddevice as sd
```

```python
l=os.listdir()
```

```python
a = l[2]
a
```

```python
a[:-4]
```

```python
samples, sr = librosa.load(a)
```

```python
D = np.abs(librosa.stft(samples))**2
```

```python
S = librosa.feature.melspectrogram(S=D)
```

```python
S = librosa.feature.melspectrogram(y=samples, sr=sr, n_mels=128, fmax=8000)
```

```python
S
```

```python
log = librosa.power_to_db(S, ref=np.max)
```

```python
plt.figure(figsize=(10, 8))
plt.subplot('211')
librosa.display.specshow(log, y_axis='mel', fmax=8000, x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Mel spectrogram')  
plt.tight_layout()
```

```python
s,r = librosa.load('Harm-c#-no-tanpura.mp3')
d1 = np.abs(librosa.stft(s))**2
s1 = librosa.feature.melspectrogram(S=d1)
s1 = librosa.feature.melspectrogram(y=s, sr=r, n_mels=128, fmax=8000)
lo = librosa.power_to_db(s1, ref=np.max)
```

```python
plt.figure(figsize=(10, 8))
plt.subplot('211')
librosa.display.specshow(lo, y_axis='mel', fmax=8000, x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Mel spectrogram2')  
plt.tight_layout()
```
